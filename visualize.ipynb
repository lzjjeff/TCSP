```python
def read_loss(file_path, tag="Validtation loss"):
    x = []
    y = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            if tag not in line:
                continue
            split1 = line.strip().split('|')
            for sp in split1:
                if 'EPOCH' in sp:
                    split2 = sp.strip().split(' ')
                    x.append(int(split2[-1]))
                elif tag in sp:
                    split2 = sp.strip().split(' ')
                    y.append(float(split2[-1]))

    assert len(x) == len(y)
    return x, y
```

```python
import matplotlib.pyplot as plt
```

```python
v_path1 = "./result/trans/lstm/translation_result_w2v.txt"
v_path2 = "./result/trans/translation_result_w2v.txt"
```

```python
x_1, y_1 = read_loss(v_path1)
x_2, y_2 = read_loss(v_path2)
plt.ylim(0.0, 0.5)
plt.title('translation v loss')
plt.plot(x_1, y_1)
plt.plot(x_2, y_2, c='red')
plt.legend(['lstm', 'transformer'])
plt.show()
```

```python
a_path1 = "./result/trans/lstm/translation_result_w2a.txt"
a_path2 = "./result/trans/translation_result_w2a.txt"
```
